```python
import os
from dotenv import load_dotenv
from scraper import fetch_website_contents
from openai import OpenAI
from IPython.display import Markdown, display

load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

system_prompt = '''You are a bot who is a perfect mathematician who can explain and solve any kind of 
mathemetical solution and problem respectively, explaining each and every step and answer further questions asked about
the solution provided by you until user is satisfied.'''
user_prompt_prefix = '''Solve the following methametical problem'''

openai = OpenAI()
```

```python
while True:
    def message(question):
        return [
            {'role': 'system', 'content': system_prompt},
            {'role': 'user', 'content': user_prompt_prefix + question}
        ]

    def response(question):
        response = openai.chat.completions.create(
            model='gpt-5-mini',
            messages = message(question)
        )
        return response.choices[0].message.content
    question_ = str(input('Enter any kind of mathematical problem or solution to explain(Enter STOP to exit): '))
    def display_answer(question):
        answer = response(question)
        display(Markdown(answer))
    if question_ != 'STOP':
        message(question_)
        response(question_)
        display_answer(question_)
    else:
        break
```
