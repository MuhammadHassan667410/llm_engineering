```python
import os
from dotenv import load_dotenv
from openai import OpenAI
import gradio as gr

load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')
openai=OpenAI(api_key=openai_api_key)
```

```python
def prompt_to_emoji(message, history):
    messages = [{"role": "system", "content": """You are a helpful assistant that translates each word in
              the English text to its corresponding emoji equivalent."""}]
    for user_msg, bot_msg in history:
        messages.append({"role": "user", "content": user_msg})
        messages.append({"role": "assistant", "content": bot_msg})
    messages.append({"role": "user", "content": f"Translate the following text to emojis: {message}"})
    
    response = openai.chat.completions.create(
        model="gpt-5",
        messages=messages,
        stream=True
    )
    result = ""
    for chunk in response:
        stream = chunk.choices[0].delta.content or ""
        result += stream
        yield result
```

```python
view = gr.ChatInterface(
    fn=prompt_to_emoji,
    title="Prompt to Emoji Translator",
    description="Enter a text prompt and get its emoji representation in real-time."
)
view.launch()
```
